```python
# base

sexo= (input('Qual seu sexo?'))

# calcular imc masculino

if sexo =='m':
  peso= float(input('qual seu peso? '))
  altura= float (input('qual sua altura? '))

  valorimc= peso/(altura**2)

  if valorimc<=20:
    print('você está abaixo do seu peso ideal')
  elif valorimc >=20 and valorimc <= 24.9:
    print('você está no seu peso ideal')
  elif valorimc>=25.0:
    print('você está acima do seu peso ideal')
  print('seu imc: ',valorimc)
  

# calcular imc feminino

elif sexo == 'f':
  peso= float(input('qual seu peso? '))
  altura= float (input('qual sua altura? '))

  valorimc= peso/(altura**2)

  if valorimc<=19:
    print('você está abaixo do seu peso ideal')
  elif valorimc >=19 and valorimc <= 23.9:
    print('você está no seu peso ideal')
  elif valorimc>=24.0:
    print('você está acima do seu peso ideal')
 
  print('seu imc: ',valorimc)
# mensagem de erro
else:
  print('Sexo invalido')





  


```
