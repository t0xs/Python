```python

# calculo imposto de renda

salario =float(input('Qual seu salário?'))
dependente = int(input('Informe seu numero de dependentes: '))

valordependente = dependente * 189.59
salario1 = salario - valordependente


if salario1 <= 1903.98:
  valorimposto = 0

if salario1>= 1903.99 and salario1 <= 2826.65:
  valorimposto = salario1 * 0.075 - 142.80

if salario1>= 2826.66 and salario1 <= 3751.05:
  valorimposto = salario1 * 0.15 - 354.80

if salario1>= 3751.06 and salario1 <= 4664.68:
  valorimposto = salario1 * 0.225 - 636.13

if salario1>4664.68 :
  valorimposto = salario1 * 0.275 - 869.36


salarioliquido = salario - valorimposto

print('seu salário é ', salario,
        'o imposto é ', valorimposto,
        'o salário líquido é ', salarioliquido)




```
